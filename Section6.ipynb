```python
work_hours = [('Abby',100), ('Billy', 400), ('Jan', 1000)]
```

```python
def employee_check(work_hours):

  current_max = 0
  employee_of_month = ''

  for employee, hours in work_hours:
    if hours > current_max:
      current_max = hours
      employee_of_month = employee
    else:
      pass

  #Return
  return (employee_of_month, current_max)
```

```python
result = employee_check(work_hours)
result
```

```python
name,hours = employee_check(work_hours)

name
```

```python
#52. Function Practice - Solutions Level One

def old_macdonald(name):

  first_letter = name[0]
  inbetween = name[1:3]
  fourth_letter = name[3]
  rest = name[4:]

  return first_letter.upper() + inbetween + fourth_letter.upper() + rest

old_macdonald('hello there')

```

```python
def blackjack(a,b,c):
  if sum([a,b,c]) <= 21:
    return sum([a,b,c])
  elif 11 in [a,b,c] and sum([a,b,c]) > 21:
    return sum([a,b,c]) - 10
  else:
    return('BUST')

blackjack(2,16,11)
```

```python
def summer_69(arr):

  total = 0
  add = True

  for num in arr:
    while add:
      if num!= 6:
        total += num
        break
    else:
      add = False
    while not add:
      if num!=9:
        break
    else:
      add = True
        break

return total
```

```python
def spy_game(nums):

  code = [0,0,7,'x']
  
  for num in nums:
    if num == code[0]:
      code.pop(0)
      
  return len(code) == 1
```

```python
spy_game([1,2,3,0,0,2,7])
```

```python
def get_change(current, previous):
    if current == previous:
        return 0
    try:
        return (abs(current - previous) / previous) * 100.0
    except ZeroDivisionError:
        return float('inf')
```

```python
get_change(80,100)
```

```python
def unique_list(list1):
  seen_numbers = []
  for number in list1:
    if number not in seen_numbers:
      seen_numbers.append(number)

  return seen_numbers

unique_list([1,2,2,3,3,3,3,4,4,5])
```

```python
def multiply(numbers):
  total = 1 
  
  # This is because 1 * n is always equal to n (e.g. 1 * 200 is equal to 200, and 1 *400 is equal to 400). 
  #If I put 0, 0 * n is alwayts equal to 0. 

  for num in numbers:
    total = total * num
  return total

multiply([1,2,3,3])
```

```python
def palindrome(s):
  # REMOVE SPACES STRING
  s = s.replace(' ', '')

  #Check is string == reversed version
  if s == s[::-1]:
    print(f'The string, "{s}" is a palindrome')
  else:
    print(f'The string, "{s}" is not a palindrome')

palindrome('nu rse s ru n')
```

```python
import string

def ispangram(str1, alphabet= string.ascii_lowercase):
  alphaset = set(alphabet)
  str1 = str1.replace(' ', '')
  str1 = str1.lower()
  str1 = set(str1)

  return str1 == alphaset

ispangram("Hello my name is Jan Kermer")
```
